```python
import numpy as np
import sys
```

```python
cost_functions_dict = {'1': 'Maximal Profit', '2':'Minimum Weight', '3':'Weight/Profit'}
knapsack_algorithms_dict = {'1': '1/0', '2':'Fractional'}
```

```python
def sort(listt, sort_on='p'):
  swapped = True
  array = listt.copy()
  n = len(array)
  index = (0 if sort_on == 'p' else 1)
  while swapped :
    swapped=False
    for i in range(n-1):
      if array[i+1][index] < array[i][index]:
        temp = array[i]
        array[i] = array[i+1]
        array[i+1] = temp
        swapped = True
    n -= 1
  return array
```

```python
def populate_elements():
  knapsack_size = input_verifier('enter the bag size: ', int, bound_check=True, upper_boundary=sys.maxsize, lower_boundary=1)
  values = input_verifier("enter elements' values space separated: ", int, single=False)
  weights = input_verifier("enter the elements' weights space separated: ", float, single=False)
  return knapsack_size,values,weights
```

```python
def type_check(input_string, iterable_type):
  try:
    val = list(map(iterable_type, input_string.split(' ')))
    return True, val
  except Exception as e:
    return False
```

```python
def boundary_check(var, low_bound, up_bound):
  return True if var>=low_bound and var<=up_bound else False
```

```python
def cost_function(choice, profits, weights):
  x = list(zip(profits, weights))
  profits = np.array(profits)
  weights = np.array(weights)
  output=[]
  if choice == 1:
    maximum_profit = sort(x)
    output.append(maximum_profit)
  elif choice == 2:
    minimum_weight = sort(x, sort_on='w')
    minimum_weight = reversed(minimum_weight) 
    output.append(list(minimum_weight))
  elif choice == 3:
    profit_weight = list((zip(x, profits/weights)))
    sorted_profit_weight = sort(profit_weight, sort_on='p/w')
    output.append(sorted_profit_weight)
  elif choice == 4:
    maximum_profit = sort(x)
    minimum_weight = sort(x, sort_on='w')
    minimum_weight = reversed(minimum_weight)
    profit_weight = list((zip(x, profits/weights)))
    profit_weight = sort(profit_weight, sort_on='p/w')
    output.append(maximum_profit)
    output.append(list(minimum_weight))
    output.append(profit_weight)
  return output
```

```python
def input_verifier(msg, var_type, single=True, bound_check=False, upper_boundary=0, lower_boundary=0):
  check=False
  while check == False:
    variable = input(msg)
    check = type_check(variable, var_type)
    if not check:
      print('error in input, if multiple elements please enter the values space separated or make sure entered values are numbers')
    elif single == True and len(check[1]) > 1:
      print('Error in input, Multiple elements entered for single variable')
      check = False
    elif single == True and bound_check == True:
      if not boundary_check(int(check[1][0]), lower_boundary, upper_boundary):
        print('Boundary error please enter value within specified range')
        check=False
  return int(check[1][0]) if single == True else check[1]
```

```python
def knapsack_run(choice, cost_choice, cost_output, knapsack_size):
  global cost_functions_dict
  global knapsack_algorithms_dict
  print(knapsack_algorithms_dict[str(choice)], ' with: ')
  if choice == 1:
    for i in range(len(cost_output)):
      if i == 2 or cost_choice == 3:
        print(' ', cost_functions_dict['3'], ':')
        knapsack_0_1(knapsack_size, cost_output[i], type=False)
      else:
        print(' ', cost_functions_dict[str(i+1)], ':')
        print(cost_output[i])
        knapsack_0_1(knapsack_size, cost_output[i])
  else:
    for i in range(len(cost_output)):
      if i == 2 or cost_choice == 3:
        print(' ', cost_functions_dict['3'], ':')
        fractional_knapsack(knapsack_size, cost_output[i], type=False)
      else:
        print(' ', cost_functions_dict[str(i+1)], ':')
        fractional_knapsack(knapsack_size, cost_output[i])
```

```python
def knapsack_0_1(W, arr, type=True):
  current_size=0
  total_profit=0
  iter=len(arr)-1
  for iter in range(len(arr)-1, -1, -1):
    current_weight = arr[iter][1] if type else arr[iter][0][1]
    if current_size+current_weight <= W:
      current_value = arr[iter][0] if type else arr[iter][0][0]
      current_size += current_weight
      total_profit += current_value
  print('   size = ', current_size)
  print('   profit = ', total_profit)
```

```python
def fractional_knapsack(W, arr, type=True):
  current_size=0
  total_profit=0
  for i in range(len(arr)-1, -1, -1):
    current_weight = arr[i][1] if type else arr[i][0][1]
    current_value = arr[i][0] if type else arr[i][0][0]
    if not current_size+current_weight < W:
      remaining = W - current_size
      ratio = remaining / current_weight
      current_weight = ratio * current_weight
      current_value = ratio * current_value
    current_size += current_weight
    total_profit += current_value    
  print('   size = ', current_size)
  print('   profit = ', total_profit)
```
