```python
import networkx
```

```python
graph={"a":["b","c","e"],"b":["a","d"],"c":["a","d","e","f"],"d":["b","c","e"],"e":["a","d","c"],"f":["c"]}
```

### **Drawing Graph**

```python
G=networkx.Graph(graph)
networkx.draw(G, with_labels=True)
```

### **Breadth first**

```python
bfs= networkx.bfs_tree(G, source='a')
networkx.draw(bfs, with_labels=True)
```

### **Depth first**

```python
dfs= networkx.dfs_tree(G, source='a')
networkx.draw(dfs, with_labels=True)
```

### **Dijkstra**

```python
g=networkx.Graph()
v1=[0,0,1,1,1,1,2,2,2,3,3,4,4,5,5,5,6,6,6,6,7,7,7,7]
v2=[1,2,3,5,6,0,0,5,6,1,7,6,7,1,2,7,1,2,4,7,3,5,6,4]
weight= [8,9,1,7,9,8,9,6,4,1,4,7,7,7,6,2,9,4,7,6,4,2,6,7]

for i in range(len(v1)):
    g.add_edge(v1[i],v2[i],weight=weight[i])

networkx.draw(g,with_labels=True)
networkx.dijkstra_path(g,0,4)
```
