### **Queue Implementation**

```python
class Queue:
  def __init__(self, passed_size=10):
    self.array=[]
    self.size=passed_size
    self.array_size = len(self.array)
  
  def push(self, arr):
    if self.size - self.array_size >= len(arr):
      self.array.extend(arr)
      self.array_size += len(arr)
    else:
      print('array is full')

  def increase_size(self, value):
    self.size = self.size + value
    
  def is_empty(self):
    return self.array_size == 0
  
  def pop(self):
    if not self.is_empty():
      element=self.array[0]
      self.array = self.array[1:]
      self.array_size -=1
      return element
    else:
      print("empty list warning")
      return None
```

### **Stack Implementation**

```python
class Stack:
  def __init__(self, passed_size=10):
    self.array=[]
    self.size=passed_size
    self.array_size = len(self.array)
  
  def push(self, arr):
    if self.size - self.array_size >= len(arr):
      self.array.extend(arr)
      self.array_size += len(arr)
    else:
      print('array is full')

  def increase_size(self, value):
    self.size = self.size + value

  def is_empty(self):
    return self.array_size == 0
  
  def pop(self):
    if not self.is_empty():
      element=self.array[-1]
      self.array = self.array[:-1]
      self.array_size -=1
      return element
    else:
      print("empty list warning")
      return None
```

### **Breadth First Iterative**

```python
def bfs_scratch(G, start):
  global Queue
  current=Queue(200)
  visited=[]
  current.push(G[start])
  vertex=start
  visited.append(vertex)
  while current.array_size >= 1:
    vertex = current.pop()
    if vertex not in visited:
      current.push(G[vertex])
      visited.append(vertex)
  print(visited)
```

### **Depth First Iterative**

```python
def dfs_scratch(G, start):
  global Stack
  current=Stack(200)
  visited=[]
  current.push(start)
  vertex=start
  while current.array_size >= 1:
    vertex = current.pop()
    if vertex not in visited:
      children = G[vertex]
      unique = [item for item in children if item not in current.array]
      current.push(unique)
      visited.append(vertex)
  print(visited)
```

### **Detph First Recursive**

```python
def dfs_recursive(G, current, visited, vertex, first=True):
  if current.array_size == 0 and first == True:
    current.push(G[vertex])
  
  if first == True:
    visited.append(vertex)
    return dfs_recursive(G, current , visited, current.pop(), False)
  
  elif vertex not in visited and first == False:
    children = G[vertex]
    unique = [item for item in children if item not in current.array]
    current.push(unique)
    visited.append(vertex)
    
    if current.array_size <= 2 and first == False:
      return visited

  return dfs_recursive(G, current , visited, current.pop(), False)
```
