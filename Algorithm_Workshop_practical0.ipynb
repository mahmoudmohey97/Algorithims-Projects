**Factorial**

```python
def fact_iterative(n):
  result=1
  for i in range(2, n+1):
    result*= i
  return result
```

```python
fact_iterative(3)
```

**Recursive Factorial**

```python
def fact_recursion(n):
  if n == 0:
    return 1
  if n == 1:
    return n
  else:
    return n*fact_recursion(n-1)
```

```python
fact_recursion(4)
```

**Fibonacci**

```python
def fib_iterative(n):
  n1,n2 = 0,1
  z=0
  for count in range(n-1):
    z = n1 + n2
    n1 = n2
    n2 = z
  return z
```

**Fibonacci Recursive**

```python
def fib_recursive(n):
  if n <= 2:
    return 1
  else:
    return fib_recursive(n-1) + fib_recursive(n-2)
```

```python
fib_iterative(8)
```

```python
fib_recursive(10)
```

**Binary Search**

```python
def search_binary(arr, target):
  index = -1
  low, high = 0,len(arr)
  mid = int(len(arr)/2)
  while high!=low :
    if arr[mid] == target:
      index = mid
      break
    elif target > arr[mid]:
      low = mid 
      high = high
    else:
      low = low
      high = mid
    mid= int((low+high) / 2)
  return index
```

**Recursive Binary Search**

```python
def search_binary_recursive(arr, mid, target, low, high):
  
  if arr[mid] == target :
    return mid
  elif low == high:
    return -1
  elif target > arr[mid]:
    mid = int((low+high)/ 2)
    return search_binary_recursive(arr, mid, target, low=mid, high=high)
  else:
    mid = int((low+high)/ 2)
    return search_binary_recursive(arr, mid, target, low=low, high=mid)
```

```python
search_binary([11, 12, 22, 25, 64], 64)
```

```python
array=[11, 12, 22, 25, 64]
low=0
high=len(array)
mid = int((low+high) / 2)
search_binary_recursive(array, mid, 22, low, high)
```

**Linear Search**

```python
def search_linear(z, target):
  for index, item in enumerate(z):
    if item == target:
      print(f'found in index: {index}')
      return index
  return -1
```

**Recursive Linear Search**

```python
def search_linear_recursion(z, target):
  if len(z) == 0 or z[len(z) - 1] == target:
    return len(z) - 1
  else:
    return search_linear_recursion(z[:-1], target)
```

```python
x = [5,4,6,1,9,8]
search_linear(x, target=6)
```

```python
search_linear_recursion(x, 0)
```

**Iterative GCD**

```python
def GCD(num1, num2):
  greatest = 0
  smallest = 0
  result=1
  if num1%num2 != num1:
    greatest = num1
    smallest = num2
  else:
    smallest = num1
    greatest = num2
  for i in range(int(smallest**(1/2)) , 1, -1):
    if greatest % i == 0 and smallest % i ==0:
      result=i
      return result
  return result
```

```python
GCD(14, 13)
```

**Ecludian GCD**

```python
def eculd_GCD(num1,num2):
  greatest = 0
  smallest = 0
  remainder = 1
  iterations=0
  if num1%num2 != num1:
    greatest = num1
    smallest = num2
  else:
    smallest = num1
    greatest = num2
  while greatest % smallest != 0:
    remainder = greatest % smallest
    greatest = smallest
    smallest = remainder
    iterations+=1
  return remainder
```

```python
eculd_GCD(14, 13)
```

**Recursive Ecludian GCD**

```python
def eclud_gcd_recursive(num1, num2):
  if num1 % num2 == 0:
    return num2
  return eclud_gcd_recursive(num2, num1%num2) 
```

```python
eclud_gcd_recursive(14, 13)
```

**Bubble Search**

```python
def bubble_sort(x):
  for i in range(len(x)-1):
    for j in range(i+1, len(x)):
      if x[j] < x[i]:
        temp=x[i]
        x[i] = x[j]
        x[j] = temp
  print(x)
```

**Optimized Bubble Sort**

```python
def bubble_sort_opt(x):
  sorted = False
  i=0
  while not sorted :
    sorted=True
    for j in range(i+1, len(x)):
      if x[j] < x[i]:
        temp=x[i]
        x[i] = x[j]
        x[j] = temp 
        sorted = False
    i+=1
  print(x)
```

```python
bubble_sort([5,1,4,2,8])
```

```python
bubble_sort_opt([5,1,4,2,8])
```

**Insertion Sort**

```python
def insertion_sort(x):
  for i in range(len(x)-1):
    for j in range(i+1, 0, -1):
      if x[j] < x[j-1]:
        temp=x[j-1]
        x[j-1] = x[j]
        x[j] = temp
  print(x)
```

```python
insertion_sort([4,3,2,1])
```

**Recursive insertion sort**

```python
def insertion_sort_recursive(x, base, index):
  if base == len(x):
    return x
  # This should behave as inner loop condition
  if index == 0:
    return insertion_sort_recursive(x, base+1,  base +1)  
    
  elif x[index] < x[index - 1]:
    x[index], x[index-1] = x[index-1], x[index]
    return insertion_sort_recursive(x, base, index - 1)
  return insertion_sort_recursive(x, base+1,  index + 1)
```

```python
insertion_sort_recursive([4,3,2,1], 0, 1)
```
