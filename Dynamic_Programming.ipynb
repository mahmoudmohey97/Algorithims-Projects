### **Knapsack DP**

```python
weight=8
profits=np.array([2,3,4,1])
weights=np.array([3,4,5,6])
items=np.zeros(weights.shape[0])
rows=weights.shape[0]
```

```python
matrix=np.zeros((rows+1,weight+1))
matrix
```

```python
for i in range(1, rows+1):
  for j in range(1, weight+1):
    weight_index= j - weights[i-1]
    if weight_index < 0:
      matrix[i,j] = matrix[i-1,j]
    else:
      matrix[i,j] = max(matrix[i-1,j], (matrix[i-1, weight_index] + profits[i-1]))
```

```python
matrix
```

```python
max_value=matrix[rows,weight]
```

```python
max_value
```

```python
j = weight
i = rows
while i != 0:        
  if matrix[i,j] != matrix[i-1, j]:
    items[i-1] = 1
    j -= weights[i - 1]
  i -= 1
```
